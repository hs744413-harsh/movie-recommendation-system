```python
features =['title' ,'genres' , 'overview' , 'production_companies' , 'vote_average','vote_count','popularity']

df = df[features]
```

```python
import ast

df['genres'] = df['genres'].apply(ast.literal_eval)
```

```python
df['genres'] = df['genres'].apply(
    lambda x: [i['name'] for i in x]
)
```

```python
df['genres'] = df['genres'].apply(lambda x: " ".join(x))
```

```python
df['genres'][0]
```

```python
df['overview'] = df['overview'].fillna(
    df['genres'] + ' movie'
)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
```

```python
df['production_companies'] = df['production_companies'].fillna('')
```

```python
df['vote_average'] = df['vote_average'].fillna(df['vote_average'].mean())
```

```python
df['vote_count'] = df['vote_count'].fillna(0)
```

```python
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
```

```python
df['popularity'][0]
```

```python
df['popularity'] = df['popularity'].fillna(df['popularity'].mean())
```

```python
df.isnull().sum()
```

```python
df['production_companies'] = df['production_companies'].apply(ast.literal_eval)
```

```python
df['production_companies'] = df['production_companies'].apply(
    lambda x: [i['name'] for i in x]
)
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: " ".join(x))
```

```python
df['production_companies'][0]
```

```python
df.head()
```

```python
df['tags'] = df[['genres', 'overview']].fillna('').agg(' '.join, axis=1)
```

```python
df.head()
```

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(txt):
    txt = str(txt).lower()
    txt = re.sub(r'[^a-zA-Z\s]','', txt)

    words = txt.split()
    # remove stopwords
    words = [word for word in words if word not in stop_words]

    #lemmatizer
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

# vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

## Model

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
    if title not in indices:
        return ['Movie not Found']
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df["title"].iloc[similar_idx]
```

```python
recommend('Queerama')
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
